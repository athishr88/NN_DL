```python
# Setting warnings off

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import warnings
warnings.filterwarnings('ignore')
```

```python
# Imports

import tensorflow as tf
import numpy as np
tf.test.gpu_device_name()
```

```python
# Data import

from sklearn.model_selection import train_test_split
from tensorflow.python.keras.datasets import cifar10

(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42)
```

```python
X_train[0].shape
```

```python
# Formatting the data

from tensorflow.keras.utils import to_categorical
X_train = X_train.reshape(X_train.shape[0], 32, 32, 3)
X_test = X_test.reshape(X_test.shape[0], 32, 32, 3)
X_val = X_val.reshape(X_val.shape[0], 32, 32, 3)

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
y_val = to_categorical(y_val)
```

```python
# Variables in the architecture
conv_layers = [2, 3, 4]
regularizers_list = ['base_model', 'l2', 'dropout', 'batchnorm']
```

```python
# Regularizers import
from tensorflow.python.keras.layers import BatchNormalization
from tensorflow.python.keras.layers.core import Dropout
from keras.regularizers import l2
```

## Model

```python
# All models tried in a 'for loop'

from tensorflow.python.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.python.keras import Sequential
from matplotlib import pyplot as plt
import time

l2_list = [None, l2(0.00001), None, None]
histories = []
times_taken = []
accuracies = []

for layer in conv_layers:
    for j, regularizer in enumerate(regularizers_list):
        
        # Building architecture
        model = Sequential()
        
        model.add(Conv2D(32, 3, activation='relu', input_shape=(32, 32, 3), kernel_regularizer=l2_list[j]))
        if regularizer == 'batchnorm':
            model.add(BatchNormalization())
        
        model.add(Conv2D(32, 3, activation='relu', kernel_regularizer=l2_list[j]))
        if regularizer == 'batchnorm':
            model.add(BatchNormalization())
        model.add(MaxPooling2D((2,2)))
        if regularizer == 'dropout':
            model.add(Dropout(0.1))
        
        if layer > 2:
            print(layer)
            model.add(Conv2D(64, 3, activation='relu', kernel_regularizer=l2_list[j]))
            if regularizer == 'batchnorm':
                model.add(BatchNormalization())
            model.add(MaxPooling2D((2,2)))
            if regularizer == 'dropout':
                model.add(Dropout(0.1))
        if layer > 3:
            model.add(Conv2D(64, 3, activation='relu', kernel_regularizer=l2_list[j]))
            if regularizer == 'batchnorm':
                model.add(BatchNormalization())
            model.add(MaxPooling2D((2,2)))
            if regularizer == 'dropout':
                model.add(Dropout(0.1))
            
        model.add(Flatten())
        model.add(Dense(100, activation='relu'))
        if regularizer == 'batchnorm':
            model.add(BatchNormalization())
        if regularizer == 'dropout':
            model.add(Dropout(0.1))
        model.add(Dense(10, activation='softmax'))
        
        #Model summary and training
        model.summary()
        opt = tf.keras.optimizers.Adam(learning_rate=0.001)
        model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
        print(f"Model --> Num conv: {layer}, Regularizer: {regularizer}")
        start_time = time.time()
        history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_data=(X_val, y_val), verbose=0)
        end_time = time.time()
        duration = end_time - start_time
        
        # Evaluation
        loss, acc = model.evaluate(X_test, y_test, verbose =0)
        
        # Bookkeeping
        times_taken.append(duration)
        print(f"Time taken to train the model = {duration:.2f} seconds")
        histories.append(history)
        accuracies.append(acc)
        print(f"Accuracy in test set : {100 * acc:.2f}")
        
        # Plotting
        plt.figure()
        plt.plot(history.history['accuracy'], color='red', label='Train set accuracy')

        plt.plot(history.history['val_accuracy'], color='blue', label='Validation set accuracies')
        plt.title("Conv Layers: %d Regularizer: %s" %(layer,regularizer))
        plt.ylabel("Accuracy")
        plt.xlabel("Number of epochs")
        
        plt.legend()
        plt.show()
            
        


```

## Best performing model

```python
# Best model

model = Sequential()

model.add(Conv2D(32, 3, activation='relu', input_shape=(32, 32, 3)))
model.add(BatchNormalization())

model.add(Conv2D(32, 3, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Dropout(0.1))

model.add(Conv2D(64, 3, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.1))
model.add(Conv2D(64, 3, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.1))

model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(Dense(10, activation='softmax'))

# Model summary and training
model.summary()
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
start_time = time.time()
history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_data=(X_val, y_val), verbose=1)
end_time = time.time()
duration = end_time - start_time

# Evaluation
loss, acc = model.evaluate(X_test, y_test, verbose =0)

print(f"Time taken to train the model = {duration:.2f} seconds")
print(f"Accuracy in test set : {100 * acc:.2f}")

# Plotting
plt.figure()
plt.plot(history.history['accuracy'], color='red', label='Train set accuracy')

plt.plot(history.history['val_accuracy'], color='blue', label='Validation set accuracies')
plt.title("Plot")
plt.ylabel("Accuracy")
plt.xlabel("Number of epochs")

plt.legend()
plt.show()


```
